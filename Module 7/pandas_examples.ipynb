```python
#we can drop columns that have one or more missing values
#inplace = True keeps us from having to create a copy
df3 = df.copy()
df3.dropna(axis=1,inplace=True)
df3.shape
```

```python
df3.head()
```

```python
#replacing null values
df_replace = df.copy()
df_replace.fillna(1000,inplace=True)
df_replace.head()
```

```python
#you can specify values for different columns that need to be replaced
df_replace2 = df.copy()
df_replace2['name'] = df_replace2['name'].fillna("SpongeBob Squarepants")
df_replace2['class'] = df_replace2['class'].fillna("Kinder")
df_replace2['mark'] = df_replace2['mark'].fillna(0)
df_replace2['gender'] = df_replace2['gender'].fillna("non-binary")
df_replace2.head()
```

```python
#renaming columns
#provide a dictionary map or assign all columns
df.rename(columns = {'id':'student_id'},inplace=True)
df.head()
```

```python
#this data set isn't great to show this but we can create pivot tables
#this setup will show us the average score by gender for each class
#we pass in -> data, values = (columns to aggregate), index -> keys to group by in table index
#columns = keys to group by on the columns
#aggfunc aggregate function to use on values.

import numpy as np
pd.pivot_table(df_replace2, values='mark',index='class',columns=['gender'],aggfunc= np.mean)
```
